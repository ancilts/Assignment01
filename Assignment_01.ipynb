##5. Visualisation

Histogram

```python
sns.histplot(df['SL'], bins=20, kde =True)
plt.title('SL Count')
plt.xlabel('SL')
plt.show()
```

Scatterplot

```python
sns.scatterplot(x='SW', y='SL' , data = df)
plt.title('Relation btwn SW & SL')
plt.show()
```

Barplot

```python
sns.countplot(x='Classification',data=df)
plt.title('Barplot of Classification')
```

```python
sns.countplot(x='SW',data=df)
plt.title('Barplot of SW')
```
